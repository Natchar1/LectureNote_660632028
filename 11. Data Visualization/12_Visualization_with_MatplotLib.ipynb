```python
crisis_data = [
    (datetime(2007, 10, 11), 'Peak of bull market'),
    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')
]

for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 75),
                xytext=(date, spx.asof(date) + 225),
                arrowprops=dict(facecolor='black', headwidth=4, width=2,
                                headlength=4),
                horizontalalignment='left', verticalalignment='top')
fig
```

We can create the cross value as the pivot table to find the relations between data in the dataframe

```python
party_counts = pd.crosstab(tips['day'], tips['size'])
party_counts
```

```python
party_pcts = party_counts.div(party_counts.sum(1), axis=0)
party_pcts
```

```python
sns.pairplot(trans_data, diag_kind='kde', plot_kws={'alpha': 0.2})
```

### Facet Grids and Categorical Data

Provide for the data set which we have additional grouping dimension

use facet to separate each group


```python
sns.catplot(x='day', y='tip_pct', hue='time', col='smoker',
               kind='bar', data=tips[tips.tip_pct < 1])
```

the `hue` parameter groupping the time in each day in the same facet

The `column` parameter spand the facet to multiple value of the smoker

instead of using hue parameter

we can create the facet for each row


```python
sns.catplot(x='day', y='tip_pct', row='time',
               col='smoker',
               kind='bar', data=tips[tips.tip_pct < 1])
```

we can also use the box plot

which show the median, quartiles, and outliner

```python
sns.catplot(x='tip_pct', y='day', kind='box',
               data=tips[tips.tip_pct < 0.5])
```

you can create more facet grid class from the seaborn documents

```python
workdata = pd.read_excel('C:/Users/Dell/Desktop/LectureNote/LectureNote_660632028/11. Data Visualization/examples/sale_data.xlsx')
workdata['Total Sale'] = workdata['Total Sale'].str.replace(',', '').astype(float)
workdata['Date'] = pd.to_datetime(workdata['Date'])
#Monthly Sale
monthly_sales = workdata.groupby(workdata['Date'].dt.to_period("M"))['Total Sale'].sum()
monthly_sales.index = monthly_sales.index.to_timestamp().month_name()
#Retail Monthly Sale
monthly_sales_retail = workdata.groupby([workdata['Date'].dt.to_period("M"), 'retail_code'])['Total Sale'].sum().reset_index()
monthly_sales_retail['Date'] = monthly_sales_retail['Date'].dt.strftime('%b')
sns.barplot(x=monthly_sales.index, y=monthly_sales.values)
sns.catplot(x='Date', y='Total Sale', col='retail_code',
            kind='bar', data=monthly_sales_retail, col_wrap=3)
```
